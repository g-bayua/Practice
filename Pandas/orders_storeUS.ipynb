```python
import pandas as pd
import numpy as py
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('orders.csv')
```

```python
##mengubah tipe data pada kolom order_date dan ship_date dari object menjadi datetime
df['order_date'] = pd.to_datetime(df['order_date'])
df['ship_date'] = pd.to_datetime(df['ship_date'])
```

```python
##mengubah tipe data kolom category menjadi category dari object
df['category'] = df['category'].astype('category')
df['subcategory'] = df['subcategory'].astype('category')
```

```python
df.info()
```

```python
##menghitung total sales dari setiap tahun
data1 = df.groupby(df['order_date'].dt.year)['sales'].sum().reset_index(name='total')
data1
```

```python
#menghitung profit dari setiap tahun
data2 = df.groupby(df['order_date'].dt.year)['profit'].sum().reset_index(name='total')
data2
```

```python
##menghitung sales berdasarkan category produk
data3 = df.groupby('category', observed=False)['sales'].sum().reset_index(name='total')
data3
```

```python
##menghitung profit berdasarkan category produk
data4 = df.groupby('category', observed=False)['profit'].sum().reset_index(name='total')
data4
```

```python
##menghitung profit berdasarkan segment
data5 = df.groupby('segment', observed=False)['profit'].sum().reset_index(name='total')
data5
```

```python
##menghitung profit berdasarkan segment di tahun 2019
data6 = pd.DataFrame(
        df[
            (df['order_date'].dt.year == 2019)
        ]
        .groupby('segment', observed=False)['profit'].sum()
        .reset_index(name='total profit 2019')
)
data6
```

```python
##menghitung jumlah customer berdasarkan customer_id bersifat uniqe
customer = df['customer_id'].nunique()
customer
```

```python
##menghitung profit berdasarkan negara bagian
profit_state = pd.DataFrame(
                df.groupby('state', observed=False)['profit']
                .sum().reset_index(name='total').sort_values(by='total', ascending=False)
)
profit_state
```

```python
##menghitung profit berdasarkan category ditahun 2019
category_profit1 = pd.DataFrame (
                    df[(df['order_date'].dt.year == 2019)]
                    .groupby('category', observed=False)['profit']
                    .sum()
                    .reset_index(name='total_2019')
                    .sort_values(by='total_2019', ascending=False)
)
category_profit1
```

```python
category_profit2 = pd.DataFrame (
                    df[(df['order_date'].dt.year == 2020)]
                    .groupby('category', observed=False)['profit']
                    .sum()
                    .reset_index(name='total_2020')
                    .sort_values(by='total_2020', ascending=False)
)
category_profit2
```

```python
merge_profit = category_profit1.merge(category_profit2, left_on='category', right_on='category')
merge_profit
```

```python
##membanding kan pendapatan berdasarkan kategori dari tahun 2019 sampai 2020
merge_profit['growth'] = merge_profit['total_2020']-merge_profit['total_2019']
merge_profit.sort_values(by=['growth'],ascending=False,inplace=True)
merge_profit
```
